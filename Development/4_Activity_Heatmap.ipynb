# Feature 4 : Activity Heatmap

## Objective

This notebook analyzes message activity across weekdays and hours to identify communication patterns in the WhatsApp group.

### Input
- parsed_chat.pkl

### Output
- Weekday × Hour Activity Heatmap

### Export
- activity_heatmap.pkl

```python
import pickle
```

```python
DATA_PATH = "../Data/"

with open(DATA_PATH + "parsed_chat.pkl", "rb") as file:
    chat_data = pickle.load(file)

print(f"Loaded {len(chat_data):,} messages successfully.")
```

```python
def calculate_activity_heatmap(chat_data):
    """
    Calculate the number of messages sent for every
    weekday and hour combination.

    Parameters:
        chat_data (list): Parsed WhatsApp chat data.

    Returns:
        dict: Nested dictionary containing weekday-wise
              hourly message counts.
    """

    weekdays = [
        "Monday",
        "Tuesday",
        "Wednesday",
        "Thursday",
        "Friday",
        "Saturday",
        "Sunday"
    ]

    activity_heatmap = {}

    for day in weekdays:

        activity_heatmap[day] = {}

        for hour in range(24):
            activity_heatmap[day][hour] = 0

    for message in chat_data:

        weekday = message["timestamp"].strftime("%A")
        hour = message["timestamp"].hour

        activity_heatmap[weekday][hour] += 1

    return activity_heatmap
```

```python
# Just To check 
activity_heatmap = calculate_activity_heatmap(chat_data)

print(activity_heatmap["Monday"])
```

```python
def display_activity_heatmap(activity_heatmap):
    """
    Display the weekday-wise hourly activity heatmap.

    Parameters:
        activity_heatmap (dict): Nested dictionary containing
                                 weekday-wise hourly message counts.

    Returns:
        None
    """

    print("=" * 170)
    print("ACTIVITY HEATMAP")
    print("=" * 170)

    print(f"{'Day':<12}", end="")

    for hour in range(24):
        print(f"{hour:>6}", end="")

    print()

    print("-" * 170)

    weekdays = [
        "Monday",
        "Tuesday",
        "Wednesday",
        "Thursday",
        "Friday",
        "Saturday",
        "Sunday"
    ]

    for day in weekdays:

        print(f"{day:<12}", end="")

        for hour in range(24):
            print(f"{activity_heatmap[day][hour]:>6}", end="")

        print()
```

```python
activity_heatmap = calculate_activity_heatmap(chat_data)

display_activity_heatmap(activity_heatmap)
```

```python
heatmap_summary = {
    "activity_heatmap": activity_heatmap
}
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

assert len(activity_heatmap) == 7

for day in activity_heatmap:

    assert len(activity_heatmap[day]) == 24

total_messages = 0

for day in activity_heatmap:

    for hour in activity_heatmap[day]:

        total_messages += activity_heatmap[day][hour]

assert total_messages == len(chat_data)

print("Feature 4 validation passed.")
```

```python
with open(DATA_PATH + "activity_heatmap.pkl", "wb") as file:
    pickle.dump(heatmap_summary, file)

print("activity_heatmap.pkl exported successfully.")
```
