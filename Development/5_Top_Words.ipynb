# Feature 5 : Top Words

## Objective

This notebook identifies the most frequently used meaningful words in the WhatsApp group.

### Input
- parsed_chat.pkl

### Output
- Top Frequent Words

### Export
- word_frequency.pkl

```python
import pickle
```

```python
DATA_PATH = "../Data/"

with open(DATA_PATH + "parsed_chat.pkl", "rb") as file:
    chat_data = pickle.load(file)

print(f"Loaded {len(chat_data):,} messages successfully.")
```

```python
STOP_WORDS = {

    "a", "an", "the", "is", "are", "am", "was", "were",

    "be", "been", "being",

    "to", "of", "in", "on", "at", "by", "for", "from",

    "with", "about", "into", "after", "before",

    "and", "or", "but",

    "this", "that", "these", "those",

    "it", "its", "it's",

    "i", "me", "my", "mine",

    "you", "your", "yours",

    "he", "him", "his",

    "she", "her", "hers",

    "they", "them", "their",

    "we", "our", "ours",

    "as", "if", "then", "than",

    "not", "no", "yes",

    "can", "could", "will", "would",

    "shall", "should",

    "do", "does", "did",

    "have", "has", "had"
}
```

```python
CUSTOM_IGNORE_WORDS = {

    "media",
    "omitted",

    "message",
    "deleted",

    "file",
    "attached",

    "image",
    "video",

    "document",

    "sticker",

    "edited"

}
```

```python
def clean_word(word):
    """
    Clean a single word by removing unwanted characters
    and filtering invalid words.

    Parameters:
        word (str): Original word extracted from a message.

    Returns:
        str: Cleaned word if valid, otherwise an empty string.
    """

    word = word.lower()

    if "http" in word or "www" in word:
        return ""

    if "@" in word:
        return ""

    cleaned_word = ""

    for character in word:

        if character.isalpha() or character.isdigit():

            cleaned_word += character

    word = cleaned_word.strip()

    if word == "":
        return ""

    if len(word) < 3:
        return ""

    if word.isdigit():
        return ""

    if word in STOP_WORDS:
        return ""

    if word in CUSTOM_IGNORE_WORDS:
        return ""

    return word
```

```python
# Test case only...

test_words = [
    "Hello!!!",
    "Project,",
    "https://google.com",
    "12345",
    "THE",
    "<media",
    "Exam...",
    "Python",
    "GroupDNA",
    "😊"
]

for word in test_words:
    print(f"{word:<25} -> '{clean_word(word)}'")
```

```python
def extract_words(chat_data):
    """
    Extract and clean all valid words from the chat.

    Parameters:
        chat_data (list): Parsed WhatsApp chat data.

    Returns:
        list: List of cleaned words.
    """

    words = []

    for message in chat_data:

        message_text = message["text"]

        split_words = message_text.split()

        for word in split_words:

            cleaned_word = clean_word(word)

            if cleaned_word != "":

                words.append(cleaned_word)

    return words
```

```python
# Onlt for testing...  

words = extract_words(chat_data)

print(f"Total Valid Words : {len(words):,}")

print("\nFirst 50 Words\n")

for word in words[:50]:
    print(word)
```

```python
def calculate_word_frequency(words):
    """
    Calculate the frequency of each word.

    Parameters:
        words (list): List of cleaned words.

    Returns:
        dict: Dictionary containing word frequencies.
    """

    word_frequency = {}

    for word in words:

        if word in word_frequency:

            word_frequency[word] += 1

        else:

            word_frequency[word] = 1

    return word_frequency
```

```python
def sort_word_frequency(word_frequency):
    """
    Sort words by descending frequency.

    Parameters:
        word_frequency (dict): Word frequency dictionary.

    Returns:
        list: List of (word, frequency) tuples.
    """

    sorted_words = sorted(
        word_frequency.items(),
        key=lambda item: item[1],
        reverse=True
    )

    return sorted_words
```

```python
words = extract_words(chat_data)

word_frequency = calculate_word_frequency(words)

top_words = sort_word_frequency(word_frequency)
```

```python
print("=" * 70)
print("TOP 20 WORDS")
print("=" * 70)

print(f"{'Rank':<6}{'Word':<25}{'Frequency':>12}")
print("-" * 70)

for rank, (word, count) in enumerate(top_words[:20], start=1):
    print(f"{rank:<6}{word:<25}{count:>12,}")
```

```python
word_summary = {
    "total_valid_words": len(words),
    "unique_words": len(word_frequency),
    "word_frequency": word_frequency,
    "top_words": top_words
}
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

assert len(words) > 0
assert len(word_frequency) > 0
assert len(top_words) == len(word_frequency)

total_frequency = sum(word_frequency.values())

assert total_frequency == len(words)

print("Feature 5 validation passed.")
```

```python
with open(DATA_PATH + "word_frequency.pkl", "wb") as file:
    pickle.dump(word_summary, file)

print("word_frequency.pkl exported successfully.")
```
