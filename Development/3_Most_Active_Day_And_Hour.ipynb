# Feature 3 : Most Active Day and Hour

## Objective

This notebook identifies the busiest day and busiest hour in the WhatsApp group based on message activity.

### Input
- parsed_chat.pkl

### Output
- Busiest Day
- Busiest Hour

### Export
- activity_summary.pkl

```python
import pickle
```

```python
DATA_PATH = "../Data/"

with open(DATA_PATH + "parsed_chat.pkl", "rb") as file:
    chat_data = pickle.load(file)

print(f"Loaded {len(chat_data):,} messages successfully.")
```

```python
def calculate_daily_activity(chat_data):
    """
    Calculate the total number of messages sent on each day.

    Parameters:
        chat_data (list): Parsed WhatsApp chat data.

    Returns:
        dict: Dictionary containing date-wise message counts.
    """

    daily_activity = {}

    for message in chat_data:

        current_date = message["timestamp"].date()

        if current_date in daily_activity:
            daily_activity[current_date] += 1
        else:
            daily_activity[current_date] = 1

    return daily_activity
```

```python
def calculate_hourly_activity(chat_data):
    """
    Calculate the total number of messages sent during each hour of the day.

    Parameters:
        chat_data (list): Parsed WhatsApp chat data.

    Returns:
        dict: Dictionary containing hour-wise message counts.
    """

    hourly_activity = {}

    for message in chat_data:

        current_hour = message["timestamp"].hour

        if current_hour in hourly_activity:
            hourly_activity[current_hour] += 1
        else:
            hourly_activity[current_hour] = 1

    return hourly_activity
```

```python
def generate_activity_summary(daily_activity, hourly_activity):
    """
    Generate a summary containing the busiest day and busiest hour.

    Parameters:
        daily_activity (dict): Date-wise message counts.
        hourly_activity (dict): Hour-wise message counts.

    Returns:
        dict: Summary of group activity.
    """

    busiest_day = max(daily_activity.items(), key=lambda item: item[1])
    busiest_hour = max(hourly_activity.items(), key=lambda item: item[1])

    total_days = len(daily_activity)

    average_messages = busiest_hour[1] / total_days

    activity_summary = {

        "daily_activity": daily_activity,

        "hourly_activity": hourly_activity,

        "busiest_day": {
            "date": busiest_day[0],
            "messages": busiest_day[1]
        },

        "busiest_hour": {
            "hour": busiest_hour[0],
            "total_messages": busiest_hour[1],
            "average_messages": average_messages
        }
    }

    return activity_summary
```

```python
daily_activity = calculate_daily_activity(chat_data)
hourly_activity = calculate_hourly_activity(chat_data)

activity_summary = generate_activity_summary(
    daily_activity,
    hourly_activity
)
```

```python
print("=" * 70)
print("MOST ACTIVE DAY & HOUR")
print("=" * 70)

print("\nBusiest Day")
print(f"Date      : {activity_summary['busiest_day']['date'].strftime('%d %B %Y')}")
print(f"Messages  : {activity_summary['busiest_day']['messages']:,}")

busiest_hour = activity_summary["busiest_hour"]

print("\nBusiest Hour")
print(f"Hour                    : {busiest_hour['hour']:02d}:00 - {busiest_hour['hour']:02d}:59")
print(f"Average Messages / Day  : {busiest_hour['average_messages']:.2f}")
print(f"Total Messages          : {busiest_hour['total_messages']:,}")
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

assert activity_summary["busiest_day"]["messages"] > 0
assert activity_summary["busiest_hour"]["total_messages"] > 0
assert 0 <= activity_summary["busiest_hour"]["hour"] <= 23

print("Feature 3 validation passed.")
```

```python
with open(DATA_PATH + "activity_summary.pkl", "wb") as file:
    pickle.dump(activity_summary, file)

print("activity_summary.pkl exported successfully.")
```
