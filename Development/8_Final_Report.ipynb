# Feature 8: Final Report

This notebook generates the final consolidated report for the WhatsApp Group Analyzer by combining the outputs of Features 1–7 into a single structured summary.

## Input Files

- parsed_chat.pkl
- group_overview.pkl
- activity_summary.pkl
- word_frequency.pkl
- response_analysis.pkl
- personality_archetypes.pkl

## Output File

- final_report.pkl

```python
import pickle
```

```python
DATA_PATH = "../Data/"

with open(DATA_PATH + "parsed_chat.pkl", "rb") as file:
    chat_data = pickle.load(file)

with open(DATA_PATH + "group_overview.pkl", "rb") as file:
    group_overview = pickle.load(file)

with open(DATA_PATH + "activity_summary.pkl", "rb") as file:
    activity_summary = pickle.load(file)

with open(DATA_PATH + "word_frequency.pkl", "rb") as file:
    word_frequency = pickle.load(file)

with open(DATA_PATH + "response_analysis.pkl", "rb") as file:
    response_analysis = pickle.load(file)

with open(DATA_PATH + "personality_archetypes.pkl", "rb") as file:
    feature7_summary = pickle.load(file)

normalized_archetype_scores = feature7_summary["normalized_scores"]
personality_archetypes = feature7_summary["assignments"]

print("All feature outputs loaded successfully.")
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

print(f"Chat Data Loaded                : {len(chat_data):,} messages")
print(f"Group Overview Loaded           : {'Yes' if group_overview else 'No'}")
print(f"Activity Summary Loaded         : {'Yes' if activity_summary else 'No'}")
print(f"Word Frequency Loaded           : {'Yes' if word_frequency else 'No'}")
print(f"Response Analysis Loaded        : {'Yes' if response_analysis else 'No'}")
print(f"Personality Archetypes Loaded   : {'Yes' if personality_archetypes else 'No'}")

print("\nAll required files loaded successfully.")
```

```python
print("=" * 70)
print("GROUP OVERVIEW")
print("=" * 70)

summary = group_overview["summary"]

print(f"Total Participants       : {summary['total_participants']:,}")
print(f"Total Messages           : {summary['total_messages']:,}")
print(f"Chat Start Date          : {summary['start_date'].strftime('%d %b %Y')}")
print(f"Chat End Date            : {summary['end_date'].strftime('%d %b %Y')}")
print(f"Duration (Days)          : {summary['duration_days']:,}")
print(f"Average Messages / Day   : {summary['average_messages_per_day']:.2f}")
```

```python
print("=" * 70)
print("PARTICIPANT STATISTICS")
print("=" * 70)

summary = group_overview["summary"]

most_active = group_overview["most_active"]
least_active = group_overview["least_active"]

print(f"Total Participants       : {summary['total_participants']:,}")

print("\nMost Active Participant")
print(f"Name                     : {most_active[0]}")
print(f"Messages                 : {most_active[1]:,}")

print("\nLeast Active Participant")
print(f"Name                     : {least_active[0]}")
print(f"Messages                 : {least_active[1]:,}")

print("\nTOP 10 CONTRIBUTORS")
print("-" * 70)
print(f"{'Rank':<6}{'Participant':<35}{'Messages'}")
print("-" * 70)

for rank, (participant, messages) in enumerate(group_overview["leaderboard"][:10], start=1):
    print(f"{rank:<6}{participant:<35}{messages:,}")
```

```python
print("=" * 70)
print("ACTIVITY SUMMARY")
print("=" * 70)

busiest_day = activity_summary["busiest_day"]
busiest_hour = activity_summary["busiest_hour"]

print(f"Busiest Day             : {busiest_day['date'].strftime('%d %b %Y')}")
print(f"Messages on Busiest Day : {busiest_day['messages']:,}")

print()

print(f"Busiest Hour            : {busiest_hour['hour']:02d}:00 - {busiest_hour['hour']:02d}:59")
print(f"Total Messages          : {busiest_hour['total_messages']:,}")
print(f"Average Messages        : {busiest_hour['average_messages']:.2f}")
```

```python
print("=" * 70)
print("WORD STATISTICS")
print("=" * 70)

print(f"Total Valid Words        : {word_frequency['total_valid_words']:,}")
print(f"Unique Words             : {word_frequency['unique_words']:,}")

print("\nTOP 10 WORDS")
print("-" * 70)
print(f"{'Rank':<6}{'Word':<20}{'Frequency'}")
print("-" * 70)

for rank, (word, frequency) in enumerate(word_frequency["top_words"][:10], start=1):
    print(f"{rank:<6}{word:<20}{frequency:,}")
```

```python
print("=" * 70)
print("RESPONSE ANALYSIS")
print("=" * 70)

average_response = sorted(
    response_analysis["average_response_time"].items(),
    key=lambda x: x[1]
)

silent_streaks = sorted(
    response_analysis["silent_streaks"].items(),
    key=lambda x: x[1]["days"],
    reverse=True
)

print("FASTEST RESPONDERS")
print("-" * 70)
print(f"{'Rank':<6}{'Participant':<35}{'Avg Response (min)'}")
print("-" * 70)

for rank, (participant, response_time) in enumerate(average_response[:10], start=1):
    print(f"{rank:<6}{participant:<35}{response_time:.2f}")

print()

print("LONGEST SILENT STREAKS")
print("-" * 70)
print(f"{'Rank':<6}{'Participant':<35}{'Days'}")
print("-" * 70)

for rank, (participant, streak) in enumerate(silent_streaks[:10], start=1):
    print(f"{rank:<6}{participant:<35}{streak['days']}")
```

```python
print("=" * 70)
print("PERSONALITY SUMMARY")
print("=" * 70)

archetype_count = {}

for details in personality_archetypes.values():
    archetype = details["Archetype"]
    archetype_count[archetype] = archetype_count.get(archetype, 0) + 1

print(f"{'Archetype':<30}{'Participants'}")
print("-" * 70)

for archetype, count in sorted(archetype_count.items()):
    print(f"{archetype:<30}{count}")
```

```python
print("=" * 70)
print("FINAL SUMMARY")
print("=" * 70)

print(f"This WhatsApp group consists of {summary['total_participants']} participants who exchanged "
      f"{summary['total_messages']:,} messages over {summary['duration_days']} days.")

print(f"The group averaged {summary['average_messages_per_day']:.2f} messages per day.")

print(f"The busiest day was {busiest_day['date'].strftime('%d %b %Y')} "
      f"with {busiest_day['messages']:,} messages.")

print(f"The most active participant was {most_active[0]} "
      f"with {most_active[1]:,} messages.")

print(f"The group vocabulary contains {word_frequency['unique_words']:,} unique words "
      f"and {word_frequency['total_valid_words']:,} valid words.")

print("Personality analysis indicates diverse communication styles among participants,")
print("with multiple personality archetypes identified across the group.")
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

print(f"Group Overview Loaded          : {'Yes' if group_overview else 'No'}")
print(f"Activity Summary Loaded        : {'Yes' if activity_summary else 'No'}")
print(f"Word Statistics Loaded         : {'Yes' if word_frequency else 'No'}")
print(f"Response Analysis Loaded       : {'Yes' if response_analysis else 'No'}")
print(f"Personality Data Loaded        : {'Yes' if personality_archetypes else 'No'}")

print()

print(f"Participants Analysed          : {summary['total_participants']}")
print(f"Messages Analysed              : {summary['total_messages']:,}")
print(f"Eligible Participants          : {len(personality_archetypes)} / {summary['total_participants']}")
print("Excluded Entries               : Meta AI and WhatsApp system events")
```

```python
final_report = {
    "group_overview": group_overview,
    "activity_summary": activity_summary,
    "word_statistics": word_frequency,
    "response_analysis": response_analysis,
    "personality_archetypes": personality_archetypes,
    "normalized_archetype_scores": normalized_archetype_scores
}

with open(DATA_PATH + "final_report.pkl", "wb") as file:
    pickle.dump(final_report, file)

print("final_report.pkl exported successfully.")
print("Feature 8 completed successfully.")
```
