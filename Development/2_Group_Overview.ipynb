# 📊 Feature 2 : Group Overview

## Objective
This notebook analyzes the overall characteristics of the WhatsApp group by computing participant statistics, message distribution, and chat timeline.

---

## Input
- ../Data/parsed_chat.pkl

## Output
- ../Data/overview.pkl

- Total Participants
- Chat Start Date
- Chat End Date

- Average Messages Per Day
- Most Active Participant
- Least Active Participant

- Top 10 Contributors

```python
import pickle
import os
from datetime import datetime
```

```python
DATA_PATH = "../Data/"

with open(DATA_PATH + "parsed_chat.pkl", "rb") as file:
    chat_data = pickle.load(file)

print(f"Parsed messages loaded successfully : {len(chat_data):,}")
```

```python
def calculate_summary(chat_data):
    """
    Calculates summary statistics of the WhatsApp chat.

    Parameters:
        chat_data (list): Parsed chat messages.

    Returns:
        dict: Summary statistics.
    """

    total_messages = len(chat_data)

    unique_participants = set()

    first_message = chat_data[0]["timestamp"]
    last_message = chat_data[0]["timestamp"]

    for message in chat_data:

        unique_participants.add(message["sender"])

        if message["timestamp"] < first_message:
            first_message = message["timestamp"]

        if message["timestamp"] > last_message:
            last_message = message["timestamp"]

    duration_days = (last_message.date() - first_message.date()).days + 1

    average_messages = total_messages / duration_days

    return {
        "total_messages": total_messages,
        "total_participants": len(unique_participants),
        "start_date": first_message,
        "end_date": last_message,
        "duration_days": duration_days,
        "average_messages_per_day": round(average_messages, 2)
    }
```

```python
summary = calculate_summary(chat_data)

summary
```

```python
def calculate_participants(chat_data):
    """
    Calculates participant-wise message statistics.

    Parameters:
        chat_data (list): Parsed chat messages.

    Returns:
        tuple:
            participant_statistics (dict)
            most_active (tuple)
            least_active (tuple)
            leaderboard (list)
    """

    participant_statistics = {}

    for message in chat_data:

        sender = message["sender"]

        participant_statistics[sender] = participant_statistics.get(sender, 0) + 1

    leaderboard = sorted(
        participant_statistics.items(),
        key=lambda participant: participant[1],
        reverse=True
    )

    most_active = leaderboard[0]
    least_active = leaderboard[-1]

    return participant_statistics, most_active, least_active, leaderboard
```

```python
participant_statistics, most_active, least_active, leaderboard = calculate_participants(chat_data)
```

```python
print("=" * 70)
print("PARTICIPANT STATISTICS")
print("=" * 70)

print(f"Total Participants : {len(participant_statistics)}")

print(f"\nMost Active Member")
print(f"Name     : {most_active[0]}")
print(f"Messages : {most_active[1]:,}")

print(f"\nLeast Active Member")
print(f"Name     : {least_active[0]}")
print(f"Messages : {least_active[1]:,}")

print("\nTOP 10 CONTRIBUTORS\n")

print(f"{'Rank':<5}{'Participant':<35}{'Messages':>12}{'Share':>12}")
print("-" * 70)

for rank, (name, count) in enumerate(leaderboard[:10], start=1):

    percentage = (count / len(chat_data)) * 100

    print(f"{rank:<5}{name:<35}{count:>12,}{percentage:>11.2f}%")
```

```python
print("=" * 70)
print("VALIDATION")
print("=" * 70)

assert len(chat_data) == summary["total_messages"]
assert len(participant_statistics) == summary["total_participants"]
assert most_active[1] >= least_active[1]

print("Feature 2 validation passed.")
```

```python
group_overview = {
    "summary": summary,
    "participant_statistics": participant_statistics,
    "most_active": most_active,
    "least_active": least_active,
    "leaderboard": leaderboard
}
```

```python
with open(DATA_PATH + "group_overview.pkl", "wb") as file:
    pickle.dump(group_overview, file)

print("group_overview.pkl exported successfully.")
```
